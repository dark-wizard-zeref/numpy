##Fancy Indexing
you can Pick elements from specific positions using a list/array of indices.

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50, 60])
print("Original:", arr)

# Pick elements at positions 0,2,4
print("Fancy indexing:", arr[[0, 2, 4]])
```

You can do this fancy indexing in 2D Matrix aswell

```python
matrix = np.arange(1, 13).reshape(3, 4)
print("Matrix:\n", matrix)
print("Select [row0,col1], [row1,col2]:", matrix[[0,1],[1,2]])
```

##Conditional Selection
Find values based on conditions or replace them.

```python
arr = np.array([5, 12, 18, 7, 30])

# Indices where arr > 10
print(np.where(arr > 10))

# Replace values: if >10 keep, else put 0
print(np.where(arr > 10, arr, 0))
```

##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

```python
arr = np.array([42, 5, 18, 23])
print("Sorted:", np.sort(arr))

idx = np.argsort(arr)
print("Indices:", idx)
print("Use indices:", arr[idx])
```

```python
mat = np.array([[3,7,1],[9,4,6]])
print(np.sort(mat, axis=1))   # sort each row
print(np.sort(mat, axis=0))   # sort each column
```

##Concatenation & Stacking

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.concatenate([a, b]))
```

Vertical & Horizontal Stacking

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print("vstack:\n", np.vstack((A,B)))   # vertical
print("hstack:\n", np.hstack((A,B)))   # horizontal
```
